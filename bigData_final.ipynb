```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
page = requests.get("https://web.saumag.edu/academics/collegesdepartments/")
soup = BeautifulSoup(page.content, 'html.parser')
print(soup)
```

```python
college_links = soup.find_all("a", title=re.compile(r"College of \w+"))
```

```python
print(college_links)
```

```python
college_info_list = []
for link in college_links:
    college_name = link.get('title')
    college_link = link.get('href')
    college_info_list.append((college_name,college_link))
```

```python
for i in college_info_list:
    print(i)
```

```python
content_list = []
for i,j in college_info_list:
    response = requests.get(j)
    page_content = response.content
    soup1 = BeautifulSoup(page_content, 'html.parser')
    content_list.append((i,soup1))
```

```python
print(content_list)
```

```python
span_elements = []
for i,j in content_list:
    span_element = j.find_all('div', class_ = "more-info")
    span_elements.append((i,span_element))
print(span_elements)
```

```python
span_texts = []
for i,j in span_elements:
    for span_element in j:
        span_text = span_element.get_text().strip()
        span_texts.append((i,span_text))
```

```python
for i in span_texts:
    print(i)
```

```python
result = []

for college, info in span_texts:
    info_lines = info.split('\n')  # Split the info string by newline character
    hours_required = info_lines[1].split(': ')[1]
    categories = info_lines[2].split(': ')[1].split(', ')
    delivery = info_lines[3].split(': ')[1].split(', ')
    categories_str = ', '.join(categories)
    delivery_str = ', '.join(delivery)

    result.append([college, hours_required, categories_str, delivery_str])

# Now 'result' is a list of lists with 5 columns
```

```python
print(result)
```

```python
import csv
with open('college_data.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['College', 'Hours Required', 'Categories', 'Delivery'])
    csvwriter.writerows(result)
```
